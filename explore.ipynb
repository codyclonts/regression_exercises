```python
def plot_categorical_and_continuous_vars(df, continuous, categorical):

    plot_list = []
    for cat in categorical:
        for cont in continuous:
            plot_list.append([cat, cont])
    
    for i in plot_list:
        plt.figure(figsize=(18, 5))
        plt.subplot(131)
        sns.boxplot(x=i[0], y=i[1], data=df)
#         sns.histplot(data = df, x = i[1], hue = 'county', element = 'step', bins =20)
        plt.subplot(132)
        sns.stripplot(x=i[0], y=i[1], data=df)
        plt.subplot(133)
        sns.violinplot(x=i[0], y=i[1], data=df)
#         sns.barplot(x=i[0], y=i[1], data=df)
        plt.show()
```

```python

def remove_outlier_IQR(df):
    Q1=df.quantile(0.25)
    Q3=df.quantile(0.75)
    IQR=Q3-Q1
    df2=df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR)))]
    return df2
```

# 8 statistical tests

Examine the relationship between age and tax value dollar count in each county.
Does the age of the home have an impact on the tax value dollar count?
Hypothesis

H0= There is no relationship between age and taxvaluedollarcnt.

H1= There is a relationship between age and taxvaluedollarcnt.

```python
# run statistical test to see if there is, in fact, a relationship between the two variables
alpha = .05
corr, p = stats.pearsonr(train.yearbuilt, train.taxvaluedollarcnt)
p< alpha, p, corr
```

```python
if p < alpha:
    print ('I reject the null hypothesis.')
else:
        print('I accept the null hypothesis.')
```
